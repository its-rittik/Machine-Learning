```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns ## make matplotlib graphs pretter
sns.set() ##autometically overwrite all the matplotlib graphs
```

```python
data = pd.read_csv(r"F:\Data Science\Machine Learning\Linear Regression\Simple Linear Regression\1.01. Simple linear regression.csv")
data
```

```python
data.describe()
```

### Create your first regression (y = b0 + b1x1)

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1,y)
plt.xlabel("SAT(independent)",fontsize = 20)
plt.ylabel("GPA(dependent)",fontsize = 20)
plt.show()
```

### Regression itself

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit() ## OLS describe modeel
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 0.275 + 0.0017*x1
fig= plt.plot(x1,yhat , lw = 1.5 , c='black',label='regression line')
plt.xlabel("SAT(independent)",fontsize = 20)
plt.ylabel("GPA(dependent)",fontsize = 20)
plt.show()
```
